```python
from PIL import Image
import numpy as np

raw_data = np.array([
    (255,   255,    255,    255,    255),
    (255,   0,      0,      0,      255),
    (255,   255,    255,    0,      255),
    (255,   255,    0,      0,      255),
    (255,   255,    0,      255,    255),
    (255,   255,    255,    255,    255)
], dtype=np.uint8)

rgb_raw_data = np.stack([raw_data] * 3, axis= - 1)
rgb_raw_data[1, 1, 0] = 255
rgb_raw_data[1, 2, 0] = 255
rgb_raw_data[1, 3, 0] = 255
rgb_raw_data[2, 3, 1] = 255
rgb_raw_data[3, 3, 1] = 255
rgb_raw_data[3, 2, 2] = 255
rgb_raw_data[4, 2, 2] = 255

img = Image.fromarray(rgb_raw_data, "RGB")

scale = 80
img = img.resize((img.width * scale, img.height * 80), Image.NEAREST)
img.show()
```
